# Example of 3D drawing in python

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
def draw_data(func, N=100, n=10):
    """func is a function z = func(x, y) to generate the data    
    N is the number of data points  
    n defines the wireframe"""
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Points
    x, y = np.random.random(N), np.random.random(N)
    z = func(x, y)    
    
    # Wireframe
    X = np.array([np.linspace(0, 1, n + 1) for k in xrange(0, n+1)])
    Y = np.array([[pt for k in xrange(0, n + 1)] for pt in np.linspace(0, 1, n + 1)])
    Z = func(X, Y)
    
    # Draw
    f = ax.scatter(x, y, z, alpha=.3)
    f = ax.plot_wireframe(X, Y, Z, rstride=1, cstride=1, alpha=.5)
    plt.xticks(np.linspace(0, 1, n+1), []), plt.yticks(np.linspace(0, 1, n+1), [])
    zmin, zmax = int(np.floor(Z.min())), int(np.ceil(Z.max()))
    f = ax.set_zticks(np.arange(zmin, zmax + 1), [])
    f = ax.set_zlim(zmin, zmax)
```

```python
draw_data(lambda x, y: 1 - 3 * x + 2 * y + 0.1)  # linear function
```

```python
draw_data(lambda x, y: 1 - 3 * x ** 2 - 2 * y ** 3 + 0.1)  # cubic function
```

```python
draw_data(lambda x, y: 1 - 3 * np.sin(3 * x) + 1 / (.2 + y) + 0.1)  # some fun function
```
